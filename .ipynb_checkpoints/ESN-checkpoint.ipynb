```python
import numpy as np
import data_handler as dh
import matplotlib.pyplot as plt
from metrics import mape
```

```python
forecast_horizon = 128
input_seq_len = 128
train_val_split = 0.8

minute_data =  'Data/dataset.csv'
daily_data = 'Data/Stocks/utx.us.txt'

handler = dh.data_handler(path = minute_data,
                          forecast_horizon = forecast_horizon,
                          input_seq_len = input_seq_len,
                          train_val_split = train_val_split,
                          backtest_windows = 4,
                          shuffled = False,
                          seed = 1)

X_train,Y_train,X_val,Y_val,X_test,Y_test,scaling_factor,scaling_bias = handler.get_data(scale = True,daily = False)
print("Data handler complete.")

```

```python
import ESN as model

class configuration():
    
    def __init__(self):
        self.n_inputs = 1
        self.n_outputs = 1
        self.n_reservoir=250
        self.spectral_radius=0.95
        self.sparsity=0
        self.noise=0.001,
        self.input_shift=None
        self.input_scaling=None
        self.teacher_forcing=True
        self.feedback_scaling=None
        self.teacher_scaling=None
        self.teacher_shift=None
        self.out_activation=np.sin
        self.inverse_out_activation=np.arcsin
        self.random_state=42
        self.silent=True
        
        
config = configuration()
model = model.ESN(config)

```

```python
err =  []
for i in range(X_train.shape[0]):
    
    X_train_temp = np.reshape(X_train[i],(forecast_horizon,1))
    Y_train_temp = np.reshape(Y_train[i],(forecast_horizon,1))
    Y_train_preds = model.fit(X_train_temp,Y_train_temp)
    err = np.append(err,mape(handler.rescale(Y_train_temp),handler.rescale(Y_train_preds)))
    
plt.plot(err)
# err = np.delete(err,np.where(np.isnan(err))[0])
# print(np.mean(err))
```

```python
err = []
for i in range(X_val.shape[0]):
    X_val_temp = np.reshape(X_val[i],(forecast_horizon,1))
    Y_val_temp = np.reshape(Y_val[i],(forecast_horizon,1))

    predictions = model.predict(X_val_temp)
    err = np.append(err,mape(handler.rescale(Y_val_temp),handler.rescale(predictions)))
```

```python
plt.plot(handler.rescale(Y_val_temp),label = "Truth")
plt.plot(handler.rescale(predictions),label = 'Pred')
plt.legend()
```
